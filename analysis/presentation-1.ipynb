```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
stable_250nm_1_pct = np.array([ (0, 90.094), (30, 87.517), (50, 84.977) ])
stable_250nm_2_pct = np.array([ (0, 99.952), (30, 87.681), (50, 84.728) ])
stable_65nm_1_pct = np.array([ (0, 44.733), (30, 41.035), (50, 35.292) ])
stable_65nm_2_pct = np.array([ (0, 45.560), (30, 39.267), (50, 38.018) ])
stable_65nm_3_pct = np.array([ (0, 48.382), (30, 43.968), (50, 38.284) ])
stable_90nm_lp_1_pct = np.array([ (0, 81.259), (50, 78.992) ])
```

```python
f = plt.subplot()
f.set_title("Power-up Bit Stability vs. Temperature")
f.set_xlabel("Temperature (C)")
f.set_ylabel("Percentage of Bits Stable")
f.set_xlim(-5, 55)
f.set_ylim(30, 95)
f.plot(stable_250nm_1_pct[:,0], stable_250nm_1_pct[:,1], label="CY-250nm-1")
f.plot(stable_250nm_2_pct[:,0], stable_250nm_1_pct[:,1], label="CY-250nm-2")
f.plot(stable_90nm_lp_1_pct[:,0], stable_90nm_lp_1_pct[:,1], label="CY-90nm-LP-1")
f.plot(stable_65nm_1_pct[:,0], stable_65nm_1_pct[:,1], label="CY-65nm-1")
f.plot(stable_65nm_2_pct[:,0], stable_65nm_1_pct[:,1], label="CY-65nm-2")
f.plot(stable_65nm_3_pct[:,0], stable_65nm_1_pct[:,1], label="CY-65nm-2")
f.legend()
# f.set_yscale('log')
```
